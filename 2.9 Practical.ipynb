```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

### Load data

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv") # the data should be in the same folder as your notebook
```

```python
data.info()
```

```python
data.head(10)
```

```python
print(data['Review'][0])
print(data['Review'][len(data)-1])
```

### Lowercase

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data.head()
```

### Stop word removal

```python
en_stopwords = stopwords.words('english')
en_stopwords.remove("not")
```

```python
f = lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)])
data['review_no_stopwords'] = data['review_lowercase'].apply(f)
```

```python
data.head()
```

```python
data['review_no_stopwords'][0]
```

### Punctuation

We want to remove punctation but we notice we have a few reviews with the symbol * instead of the word star. We want to keep this as it might add meaning to the review. We can do this using what we learned about regular expressions.

```python
f = lambda x: re.sub(r"[*]", "star", x['review_no_stopwords'])
data['review_no_stopwords_no_punct'] = data.apply(f, axis=1)

index_f = lambda x: re.sub(r"[*]", "star", x)
data['index_test'] = data['review_no_stopwords'].apply(index_f)
```

```python
data.head()
```

```python
f = lambda x: re.sub(r"[^\w\s]", "", x['review_no_stopwords_no_punct'])
data['review_no_stopwords_no_punct'] = data.apply(f, axis=1)

```

```python
data.head()
```

### Tokenizing

```python
f = lambda x: word_tokenize(x['review_no_stopwords_no_punct'])
data['tokenized'] = data.apply(f, axis=1)
```

```python
data.head()
```

```python
data['tokenized'][0]
```

### Stemming

```python
ps = PorterStemmer()
```

```python
f = lambda tokens: [ps.stem(token) for token in tokens]
data["stemmed"] = data["tokenized"].apply(f)
```

```python
data.head()
```

```python
data['stemmed'][0]
```

### Lemmatization

```python
lemmatizer = WordNetLemmatizer()
```

```python
f = lambda tokens: [lemmatizer.lemmatize(token) for token in tokens]
data["lemmatized"] = data["tokenized"].apply(f)
```

```python
data['lemmatized'][0]
```

```python
data.head()
```

## N-grams

```python
tokens_clean = sum(data['lemmatized'], [])
print(len(tokens_clean), tokens_clean[:20])
```

```python
# unigrams: n=1
unigrams = (pd.Series \
            (nltk.ngrams(tokens_clean, 1)) \
            .value_counts()) 
print(unigrams)
```

```python
# bigrams: n=2
bigrams = (pd.Series \
           (nltk.ngrams(tokens_clean, 2)) \
           .value_counts()) 
print(bigrams)
```
